### 1 Grafik Batang

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('dataset-reviews.csv')

data_bar = data['rating'].value_counts()
plt.bar(x=data_bar.index, height=data_bar.values)
```

### Ksimpulan 
### 1 Distribusi Rating: Grafik batang ini memberikan gambaran visual tentang bagaimana ulasan-ulasan dalam dataset ini terdistribusi berdasarkan nilai rating. Dalam hal ini, sumbu x (nilai rating) menunjukkan skala rating, dan sumbu y (jumlah ulasan) menunjukkan berapa kali setiap nilai rating tersebut muncul dalam dataset.

### 2 Rating Tertinggi: Batang dengan nilai tertinggi pada grafik ini mewakili nilai rating yang paling sering diberikan oleh pengguna dalam dataset tersebut. Semakin tinggi batang, semakin sering rating tersebut muncul.

### 3 Rating Terendah: Batang dengan nilai terendah pada grafik ini mewakili nilai rating yang jarang diberikan oleh pengguna dalam dataset tersebut. Semakin rendah batang, semakin jarang rating tersebut muncul.

### 4 Kesimpulan Distribusi: Dengan melihat grafik ini, Anda dapat dengan cepat melihat distribusi ulasan berdasarkan nilai rating, yang dapat memberikan wawasan tentang seberapa baik atau buruk produk atau layanan tersebut berdasarkan pendapat pengguna.

### 3 Piecart

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('dataset-reviews.csv')

plt.figure(figsize=(6, 6))
data['rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribusi Data berdasarkan Rating')
plt.xlabel('Rating')
plt.ylabel('Persentase')
plt.show()
```

### Kesimpulan
### 1 Diagram Lingkaran Distribusi Rating: Diagram lingkaran ini menunjukkan bagaimana persentase ulasan dalam dataset terbagi berdasarkan nilai rating. Setiap sektor di dalam diagram lingkaran mewakili satu nilai rating, dan ukuran sektor tersebut mencerminkan persentase ulasan yang memiliki rating tersebut.

### 2 Ukuran Gambar: Ukuran gambar diagram lingkaran diatur menggunakan plt.figure(figsize=(6, 6)), sehingga memiliki dimensi yang lebih kecil dan kompak.

### 3 Label dan Judul: Diagram lingkaran dilengkapi dengan judul ("Distribusi Data berdasarkan Rating") dan label sumbu x dan y yang menjelaskan bahwa sumbu x adalah nilai rating dan sumbu y adalah persentase.

### 4 Persentase: Pada setiap sektor dalam diagram lingkaran, persentase ulasan yang memiliki rating tertentu ditampilkan menggunakan autopct='%1.1f%%', sehingga Anda dapat melihat persentase dengan tepat.

### 4 Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('dataset-reviews.csv')

# Hitung korelasi antar fitur dalam dataset
correlation_matrix = data.corr()

# Buat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi Antar Fitur')
plt.show()
```
